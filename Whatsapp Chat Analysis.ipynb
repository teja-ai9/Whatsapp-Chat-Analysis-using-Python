Exploration1: Let us find all the conversations in the whatsapp group i.e without timestamps, 'joined' 'removed', etc.

```python
chat= open("chat.txt",encoding = 'utf8')

L=[]
conv=[]

for line in chat:                     #get access to each line in file
    line=line.strip()   
    pos1= line.find("-") # removing the left and right spaces
    first=line[pos1:]     #finding first position after "-"
    L.append(first)  
    
for line in L:
    if (":") in line:
        conv.append(line)
        
for i in conv:
    print(i)
    
```

Exploration2: Let us find all the locations of members in the whatsapp group.

```python
chat= open("chat.txt",encoding = 'utf8')

L=[]
conv=[]
loc=[]

for line in chat:                     #get access to each line in file
    line=line.strip()   
    pos1= line.find("-") # removing the left and right spaces
    first=line[pos1:]     #finding first position after "-"
    L.append(first)  
    
for line in L:
    if (":") in line:
        conv.append(line)
        
for line in conv:
    if ("from") in line:
        atpos = line.find ('from')   #finding @
        sppos = line.find (' ',atpos+5)  #finding space after atpos
        add = line[atpos+4:sppos]
        print(add)
```

Exploration3: Let us find out the list of unique members who have interacted atleast once in the whatsapp group.

```python
chat= open("chat.txt",encoding = 'utf8')

L=[]
members=[]

for line in chat:                     #get access to each line in file
    line=line.strip()   
    pos1= line.find("-") # removing the left and right spaces
    first=line[pos1:]     #finding first position after "-"
    L.append(first)  
    
for line in L:
    if (":") in line:
        sppos = line.find (':')  #finding space after atpos
        add = line[2:sppos]
        if add not in members:
            members.append(add)
            
for i in members:
    print(i)

```

Exploration 4: Filtering out the chats based on date.

```python
def date(chat, date):
    chat= open("chat.txt",encoding = 'utf8')
    
    for line in chat:
        if line.startswith(date):
            line= line.strip()
            pos1= line.find(date)
            add= line[pos1:]
            print(add)
```

Exploration 5: Filtering out the chats based on a person.

```python
def person(chat,person):
    chat= open("chat.txt", encoding="utf8")
    
    L=[]
    for line in chat:
        line=line.strip()
        pos1= line.find('-')
        add= line[pos1+1:]
        L.append(add)
    
    for line in L:
        if person in line:
            line= line.strip()
            pos1= line.find(person)
            pos2= line.find(':',pos1)
            add= line[pos1:]
            print(add)
```

Exploration 6: Finding and Visualising the time(AM, PM) distribution of chats.

```python
file= open("chat.txt",encoding = 'utf8')
count_am = 0
count_pm=0     #initialising counters as lists
l=[]

for line in file:                          #get access to each line in file
    if "/22" in line:                      #taking lines with date
        line = line.rstrip()               # removing the left and right spaces by strip
        #print(line)
        stpos = line.find (',')   
        enpos = line.find (':',stpos)   #setting the start and end positions
        add = line[stpos+2:enpos]
        l.append(add)                   #adding the hour values to list
            
       
           
for k in l:
    
    if (int(k)<12):                         #finding the AM hours (below 12PM)
        count_am= count_am+1
        
    else:                                   #finding the PM hours (after 12PM)
        count_pm= count_pm+1
    
print("The AM and PM hours are",count_am,",",count_pm)  #final count of AM and PM hours
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([count_am,count_pm])
labels= ["AM","PM"]

plt.pie(y, labels=labels)
plt.show() 
```
